# **Target Trial Emulation (TTE) with Clustering Analysis**

## **1Ô∏è‚É£ Introduction**
Target Trial Emulation (TTE) is a methodological framework in epidemiology that aims to **reduce biases** in traditional observational studies by simulating a **randomized clinical trial**. In this notebook, we extend TTE by integrating **K-Means clustering**, allowing us to group participants based on survival patterns.

This study is based on data from **[RPubs: TTE Simulation](https://rpubs.com/alanyang0924/TTE)**, which provides a dummy dataset (`data_censored.csv`). We first **converted the original R code into Python** and then applied clustering techniques to uncover meaningful survival trends.

- Clusters were analyzed to uncover differences in **survival outcomes**.

---

## üìä Exploratory Data Analysis (EDA)
Before applying clustering, we first analyze the data distribution. EDA allows us to:
- Identify trends in age and survival periods.
- Detect missing values or anomalies that could impact clustering.
- Ensure the dataset is suitable for unsupervised learning techniques like K-Means.



```python
import seaborn as sns
import matplotlib.pyplot as plt

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset ONCE at the start
df = pd.read_csv("csv/data_censored.csv")

# Preview data to confirm it loaded correctly
print(df.head())  
print("Columns in df:", df.columns)
df.columns = df.columns.str.strip() 

df_filtered = df[['age', 'period']].dropna()

# Summary statistics
print(df.describe())

# Distribution of Age & Survival Period
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.histplot(df['age'], bins=20, kde=True)
plt.title("Distribution of Age")

plt.subplot(1,2,2)
sns.histplot(df['period'], bins=20, kde=True)
plt.title("Distribution of Survival Period")

plt.show()
```

## üìà Elbow Method: Finding the Optimal Number of Clusters
The Elbow Method helps us determine the best value for `k` in K-Means clustering.
- The "elbow point" represents the optimal number of clusters.
- In our case, `k = 5` was chosen based on the point where inertia stops decreasing significantly.


```python
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans

# Prevent memory leak spam
import os
os.environ["OMP_NUM_THREADS"] = "1"

# Run Elbow Method
visualizer = KElbowVisualizer(KMeans(), k=(2,10))
visualizer.fit(df_filtered)
visualizer.show()

```

## üèÜ Why Use K-Means for This Dataset?
Before applying K-Means clustering, we considered different methods. K-Means was chosen because it is **well-suited for structured, well-separated clusters**, which aligns with the characteristics of our dataset.

### ‚úÖ **Why K-Means Works Best Here**
- **Well-Separated Clusters:** The dataset has distinct patient groups, making K-Means a strong choice.  
- **Clear Interpretability:** K-Means assigns each patient to a specific group, making it easier to analyze survival trends.  
- **Scalability & Efficiency:** K-Means handles large datasets efficiently, which is useful for medical data analysis.

### ‚ùå **Why We Did Not Use Other Methods**
- **DBSCAN (Density-Based Clustering)**:  
  - Works better with **noisy data and arbitrary-shaped clusters**, but our dataset lacks significant noise.  
  - Struggles when cluster densities are similar, as seen in our data.  
- **GMM (Gaussian Mixture Models)**:  
  - Assumes clusters follow a **Gaussian (normal) distribution**, which **does not match our survival period trends**.  
  - Computationally heavier and less interpretable than K-Means for this study.

By using **K-Means**, we efficiently grouped patients into clusters based on **age and survival period**, helping us analyze **risk factors and survival trends** in Target Trial Emulation.

## **3Ô∏è‚É£ Why This Matters for Target Trial Emulation (TTE)?**  

üìå **Target Trial Emulation (TTE)** aims to **reduce biases in traditional epidemiological studies**.  
- Many clinical trials fail to **properly compare treatment effects** because patients have different baseline risks.  
- By clustering patients **before analysis**, we create **risk-adjusted groups** that improve the reliability of trial outcomes.  

---

## **4Ô∏è‚É£ Practical Application: How Clustering Improves Medical Research**  

1Ô∏è‚É£ **Clustering prevents biased treatment comparisons**  
   - If we compare all patients without clustering, results may be **skewed by age or underlying risk factors**.  
   - By analyzing treatments **within clusters**, we ensure that the survival differences are due to the treatment‚Äînot external variables.  

2Ô∏è‚É£ **Personalized medicine & risk stratification**  
   - We can **identify high-risk clusters** (like Cluster 1 & 3) and prioritize them for **early intervention strategies**.  
   - Conversely, **Cluster 2 (older but long survival) may represent a group that responds well to certain treatments.**  

3Ô∏è‚É£ **Improving future clinical trials**  
   - Instead of randomly assigning patients to treatments, we could **assign treatments based on cluster characteristics**.  
   - This approach could **increase trial efficiency and accuracy**.
